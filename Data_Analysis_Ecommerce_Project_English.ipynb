## Value Grouping

### Status

```python
df['Status'].unique()
```

The status column has many unique values. we will group the status column values. There are 15 types of status which we will group into four groups, namely Complete, Canceled, In Process, and Fraud.

The value included in the **Complete** group is **complete**

The values included in the **Cancel** group are **canceled, order_refunded, refund, and closed**

The values included in the **In Process** group are **received, held, exchange, pending_paypal, paid, cod, pending, processing, and payment_review**

The value that is included in the **Fraud** group is **fraud**

```python
df['Status'] = df['Status'].replace('complete','Complete')
```

```python
df['Status'] = df['Status'].replace('canceled','Cancel').replace('order_refunded','Cancel').replace('refund','Cancel').replace('closed','Cancel')
```

```python
df['Status'] = df['Status'].replace('received','In Process').replace('holded','In Process').replace('exchange','In Process').replace('pending_paypal','In Process').replace('paid','In Process').replace('cod','In Process').replace('pending','In Process').replace('processing','In Process').replace('payment_review','In Process')
```

```python
df['Status'] = df['Status'].replace('fraud','Fraud')
```

### Income Analysis

We will first define income and losses.


* Income is the total of grand_total which has the status "complete".
* Lost revenue is the total of grand_total which has the status "cancel".
* Potential additional income is the total of grand_total which has the status "in process".
* Total revenue is Revenue + Potential Revenue (assuming all transactions will be "Complete") - Lost Revenue

```python
# Income
pendapatan = df.loc[(df["Status"] == "Complete"), "GrandTotal"].sum()

print("Pendapatan yang didapat sebanyak:", pendapatan)
```

```python
# Lost Revenua
pendapatan_lost = df.loc[(df["Status"] == "Cancel"), "GrandTotal"].sum()

print("Pendapatan yang hilang sebanyak:", pendapatan_lost)
```

```python
# Potential Additional Value
pendapatan_potensi = df.loc[(df["Status"] == "In Process"), "GrandTotal"].sum()

print("Potensi pendapatan yang akan didapat sebanyak:", pendapatan_potensi)
```

```python
# Total Revenue
total_pendapatan = pendapatan + pendapatan_potensi - pendapatan_lost
print("Total pendapatan sebanyak:", total_pendapatan)
```

# Conclusion & Recommendation

## Conclusion

1. Digital payment methods are the payment methods most frequently chosen by consumers, accounting for 53.47% of total transactions. However, this method is also the method with the most cancel status of all transactions. The large number of cancellations in this transaction caused the company to lose revenue of 1,766,662,405.16.

2. Meanwhile, the cash payment method is the second option for consumers for 46.52% of total transactions. Even though there are fewer than digital payment methods, this method has the highest complete status of all transactions. With so many complete statuses, the company earned revenue of 47,724,235.91.

3. Digital payment methods are most widely used by consumers when they want to buy goods in the Mobiles & Tablets category. Meanwhile, the cash payment method is most widely used by consumers when they want to buy goods in the Men's Fashion category.

## Recommendation

DIGITAL PAYMENTS

1. There are several possibilities that cause many transactions to be canceled using digital payment methods, such as consumers taking too long to complete payments, failures in bank or financial service provider payment systems, or a digital payment ecosystem that does not provide many choices. (source: razorpay.com)

2. To overcome this problem, companies can provide several solutions such as providing more digital payment options such as credit cards, e-wallets and local bank transfers. Apart from that, the company can also provide suggestions for the same payment method from previous successful transactions.

CASH PAYMENT

1. Cancel status on cash transactions can be caused by several things, such as the consumer's absence when the courier arrives at the location to make the payment or the consumer does not provide money according to the nominal payment.

2. To overcome this problem, companies can propose an agreement with consumers regarding the payment date that will be made when the consumer checks out the goods. Apart from that, companies can also provide cash payment limits to consumers by looking at the average successful transactions using the cash method. (source: pakistantoday.com.pk)

By providing solutions to existing problems, companies can optimize business activities with consumers. Moreover, with the Pakistani government currently encouraging business digitalization, companies can also support the government by continuing to optimize their services.
